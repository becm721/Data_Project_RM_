```python
def acts_compare():
    year = ('2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019')
    y_pos = np.arange(len(year))
    data1 = (107706, 111910, 116502, 109381, 116055, 120526, 124492, 128397, 128777, 126274, 122001, 116924)
    data2 = (28535,	28346,	28390,	26375,	26839,	27800,	27784,	29027,	29496,	29698,	31376,	32122)

    fig, ax1 = plt.subplots()

    color = 'tab:blue'
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Total No. Recorded', color=color)
    ax1.plot(year, data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx() 
    color = 'tab:red'
    ax2.set_ylabel('Acts intended to record injury No. Recoreded', color=color) 
    ax2.plot(year, data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout() 
    plt.show()
```
